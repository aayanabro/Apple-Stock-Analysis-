```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

Datetime index And Resampleing


```python
df = pd.read_csv(r'C:\Users\Latitude\Downloads\aapl.csv',parse_dates= ["Date"],index_col=["Date"])
```

```python
df.head(
)
```

```python
#Find the Apple Average Stock in Jan,2017 
df.loc["2017-01"].Close.mean()
```

```python
df=df.sort_index()
```

```python
df["2017-01-01":"2017-01-07"]
```

```python
#Resampling
#weekly stock of Apple 
df.Close.resample('W').mean().plot(kind="line")

plt.show()
```

```python
# monthly stocks of Apple
df.Close.resample('ME').mean().plot(kind="line")
plt.show()
```

```python
df.Close.resample("QE").mean().plot(kind='bar')
plt.show()
```
